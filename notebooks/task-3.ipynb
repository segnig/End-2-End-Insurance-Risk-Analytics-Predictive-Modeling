# **A/B Hypothesis Testing Analysis**

Risk Metrics:
* Claim Frequency: Proportion of policies with at least one claim
* Claim Severity: Average claim amount when a claim occurs

Profit Metric:
* Margin: TotalPremium - TotalClaims

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sys
import os

# Configure warnings (suppress non-critical warnings)
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

sys.path.append(os.path.abspath("../scripts"))

from hypothesis_testing import *
```

```python

analyzer = InsuranceRiskAnalyzer(file_path="../data/data.csv")

if not analyzer.df.empty:
    print(f"Loaded {len(analyzer.df):,} policies")
    
    # Run all tests with error handling
    analysis = analyzer.run_full_analysis()
    
    # Print reports
    for test_name, report in analysis['reports'].items():
        print(f"\n{test_name.replace('_', ' ').title()} Report:")
        print(report)
    
    # Save detailed results
    with open("insurance_risk_analysis_report.txt", "w") as f:
        for test_name, report in analysis['reports'].items():
            f.write(f"\n{test_name.replace('_', ' ').title()} Report:\n")
            f.write(report + "\n")
else:
    print("Failed to load data. Please check the file path.")
```
